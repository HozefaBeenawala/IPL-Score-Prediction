```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
df = pd.read_csv('ipldata.csv')
df.head()
```

```python
data = pd.read_csv('IPL Player Stats - 2016 till 2019.csv')
data.head()
```

```python
df= df.drop(['Unnamed: 0','extras','match_id', 'runs_off_bat'],axis = 1)
new_df = pd.merge(df,data,left_on='striker',right_on='Player',how='left')
new_df.drop(['wicket_type', 'player_dismissed'],axis=1,inplace=True)
new_df.columns
```

```python
new_df.info()
```

```python
new_df.isnull().sum()
```

```python
str_cols = new_df.columns[new_df.dtypes==object]
new_df[str_cols] = new_df[str_cols].fillna('.')
```

```python
listf = []
  
for i in new_df.columns:
    if new_df[i].dtype==object:
        print(i,"->" ,new_df[i].dtype)
        listf.append(i)
```

```python
a1 = new_df['venue'].unique()
a2 = new_df['batting_team'].unique()
a3 = new_df['bowling_team'].unique()
a4 = new_df['striker'].unique()
a5 = new_df['bowler'].unique()
  
def labelEncoding(data):
    dataset = pd.DataFrame(data)
    feature_dict ={}
      
    for feature in dataset:
        if dataset[feature].dtype==object:
            le = preprocessing.LabelEncoder()
            fs = dataset[feature].unique()
            le.fit(fs)
            dataset[feature] = le.transform(dataset[feature])
            feature_dict[feature] = le
              
    return dataset
  
labelEncoding(new_df)
```

```python
ip_dataset = new_df[['venue','innings', 'batting_team', 
                      'bowling_team', 'striker', 'non_striker',
                      'bowler']]
  
b1 = ip_dataset['venue'].unique()
b2 = ip_dataset['batting_team'].unique()
b3 = ip_dataset['bowling_team'].unique()
b4 = ip_dataset['striker'].unique()
b5 = ip_dataset['bowler'].unique()
new_df.fillna(0,inplace=True)
  
features={}
  
for i in range(len(a1)):
    features[a1[i]]=b1[i]
for i in range(len(a2)):
    features[a2[i]]=b2[i]
for i in range(len(a3)):
    features[a3[i]]=b3[i]
for i in range(len(a4)):
    features[a4[i]]=b4[i]
for i in range(len(a5)):
    features[a5[i]]=b5[i]
      
features
```

```python
X = new_df[['venue', 'innings','batting_team',
             'bowling_team', 'striker','bowler']].values
y = new_df['y'].values
  
from sklearn.model_selection import train_test_split
  
X_train, X_test, y_train, y_test = train_test_split(
  X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
  
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
  
model.add(Dense(43, activation='relu'))
model.add(Dropout(0.5))
  
model.add(Dense(22, activation='relu'))
model.add(Dropout(0.5))
  
model.add(Dense(11, activation='relu'))
model.add(Dropout(0.5))
  
model.add(Dense(1))
  
model.compile(optimizer='adam', loss='mse')
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', patience=10)
```

```python
model.fit(x=X_train, y=y_train, epochs=400, 
          validation_data=(X_test,y_test),
          callbacks=[early_stop] )
```

```python
model_losses = pd.DataFrame(model.history.history)
model_losses.plot()
```

```python
predictions = model.predict(X_test)
sample = pd.DataFrame(predictions,columns=['Predict'])
sample['Actual']=y_test
sample.head(10)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
  
mean_absolute_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```
